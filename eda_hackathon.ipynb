```python
import pandas as pd
import numpy as np
from typing import Optional, Tuple, List, Dict

# =========================================================
# 0) Utilidades de coordenadas (StatsBomb-like 120x80)
# =========================================================
def _extract_xyz(val) -> Tuple[Optional[float], Optional[float], Optional[float]]:
    if isinstance(val, (list, tuple)) and len(val) >= 2:
        x = float(val[0]) if val[0] is not None else None
        y = float(val[1]) if val[1] is not None else None
        z = float(val[2]) if (len(val) >= 3 and val[2] is not None) else None
        return x, y, z
    return None, None, None

def add_xyz_columns(df: pd.DataFrame, src_col: str, prefix: Optional[str] = None) -> pd.DataFrame:
    if src_col not in df.columns:
        return df
    if prefix is None:
        prefix = (src_col
                  .replace("_end_location", "_end")
                  .replace("location", "loc")
                  .replace("__", "_"))
    xs, ys, zs = zip(*df[src_col].apply(_extract_xyz))
    df[f"{prefix}_x"] = xs
    df[f"{prefix}_y"] = ys
    df[f"{prefix}_z"] = zs
    return df

def add_xy_all(data_all: pd.DataFrame, extra_end_cols: Optional[List[str]] = None) -> pd.DataFrame:
    df = data_all.copy()
    location_cols = [
        "location",
        "pass_end_location",
        "shot_end_location",
        "carry_end_location",
        "goalkeeper_end_location",
    ]
    if extra_end_cols:
        location_cols.extend(extra_end_cols)
    for c in location_cols:
        if c in df.columns:
            pref = (c.replace("_end_location", "_end")
                      .replace("location", "loc")
                      .replace("__", "_"))
            add_xyz_columns(df, c, prefix=pref)
    return df


def convert_pitch_units(df_with_xy: pd.DataFrame,
                        in_dims: Tuple[float, float]=(120.0, 80.0),
                        out_dims: Tuple[float, float]=(105.0, 68.0),
                        prefixes: Optional[List[str]]=None) -> pd.DataFrame:
    """
    Convert StatsBomb coordinates (120x80) to meters (default 105x68).
    Scales all '*_x'/'*_y' columns matching given prefixes; if prefixes=None, auto-detects.
    """
    df = df_with_xy.copy()
    in_x, in_y = in_dims
    out_x, out_y = out_dims
    sx = out_x / in_x
    sy = out_y / in_y

    if prefixes is None:
        prefixes = sorted(set(c[:-2] for c in df.columns if c.endswith("_x")))

    for pref in prefixes:
        xcol, ycol = f"{pref}_x", f"{pref}_y"
        if xcol in df.columns:
            df[xcol] = pd.to_numeric(df[xcol], errors="coerce") * sx
        if ycol in df.columns:
            df[ycol] = pd.to_numeric(df[ycol], errors="coerce") * sy
    return df

# =========================================================
# 1) Season normalization & helpers
# =========================================================
def _ensure_season(df: pd.DataFrame) -> pd.DataFrame:
    """
    Ensure a 'season' column exists. If missing/empty, derive from year(match_date).
    Accepts match_date as 'YYYY-MM-DD' or parseable timestamp-like.
    """
    out = df.copy()
    if "season" not in out.columns or out["season"].isna().all():
        if "match_date" in out.columns:
            years = pd.to_datetime(out["match_date"], errors="coerce").dt.year.astype("Int64")
            out["season"] = years.astype(str)
        else:
            out["season"] = "unknown"
    return out

def _safe_bool_series(s: pd.Series, default: bool=False) -> pd.Series:
    """Coerce arbitrary series into a boolean series safely."""
    if s is None:
        return pd.Series([default]*0, dtype=bool)
    if s.dtype == bool:
        return s.fillna(default)
    return s.astype(str).str.lower().isin(["true","t","1","yes","y"]).fillna(default)

# =========================================================
# 2) Minutes played (match-level) heuristic without lineups
# =========================================================
def estimate_minutes_by_match_player(data_all: pd.DataFrame) -> pd.DataFrame:
    """
    Heuristic using events only (no lineups):
      - For each match: duration ~ max(minute + second/60), clipped to [0,130].
      - For each player in that match:
            start_min = min event minute for that player
            end_min   = first event with player_off_permanent==True OR match_duration
            minutes   = max(0, end_min - start_min)
      - If minutes = 0 but player appeared, set to 1.0 to avoid divide-by-zero later.
    Returns: match_id, season, player_id, player, team, minutes_played
    """
    df = _ensure_season(data_all).copy()

    minute = pd.to_numeric(df.get("minute", np.nan), errors="coerce")
    second = pd.to_numeric(df.get("second", 0), errors="coerce").fillna(0)
    tmin = minute.fillna(0) + (second.fillna(0) / 60.0)
    df["_tmin"] = tmin

    match_dur = df.groupby("match_id")["_tmin"].max().clip(lower=0, upper=130).rename("match_duration_min")
    start_min = df.groupby(["match_id","player_id"])["_tmin"].min().rename("start_min")

    off_perm = _safe_bool_series(df.get("player_off_permanent", pd.Series(False, index=df.index)), default=False)
    df["_off_perm"] = off_perm
    off_times = df.loc[df["_off_perm"]].groupby(["match_id","player_id"])["_tmin"].min().rename("off_min")

    mp = (start_min.to_frame()
                 .join(off_times, how="left")
                 .join(match_dur, on="match_id", how="left")
                 .reset_index())

    mp["end_min"] = np.where(mp["off_min"].notna(), mp["off_min"], mp["match_duration_min"])
    mp["minutes_played"] = (mp["end_min"] - mp["start_min"]).clip(lower=0)
    mp.loc[(mp["minutes_played"] == 0) & mp["start_min"].notna(), "minutes_played"] = 1.0

    meta_cols = ["player_id","player","team","season","match_id"]
    meta = (_ensure_season(df)[meta_cols]
            .drop_duplicates(subset=["match_id","player_id"])
            .set_index(["match_id","player_id"]))
    mp = (mp.set_index(["match_id","player_id"])
            .join(meta, how="left")
            .reset_index())

    mp = mp[["match_id","season","player_id","player","team","minutes_played"]].fillna({"player":"", "team":""})
    return mp

# =========================================================
# 3) Season-team-player aggregations (totals)
# =========================================================
def aggregate_player_team_season_counts(data_all: pd.DataFrame) -> pd.DataFrame:
    """
    Aggregate by (season, team, player_id, player):
      shots, goals, xg, passes_att, passes_cmp, pass_pct, key_passes, assists, xA,
      dribbles_cmp, duels_won, interceptions, blocks, clearances,
      obv_total, obv_for, obv_against
    """
    df = _ensure_season(data_all).copy()
    for col in ["type", "player", "player_id", "team", "season"]:
        if col not in df.columns:
            df[col] = np.nan

    grp = ["season","team","player_id","player"]

    # flags (0/1) to avoid Series vs DataFrame groupby issues
    df["is_shot"] = (df["type"] == "Shot").astype(int)
    df["is_goal"] = ((df["type"] == "Shot") & (df.get("shot_outcome", pd.Series([None]*len(df))) == "Goal")).astype(int)

    xg_col = "shot_statsbomb_xg"
    df["_xg"] = np.where(df["type"].eq("Shot"), df.get(xg_col, np.nan), np.nan)

    df["is_pass"] = (df["type"] == "Pass").astype(int)
    pass_outcome = df.get("pass_outcome", pd.Series([np.nan]*len(df)))
    df["is_pass_cmp"] = ((df["type"] == "Pass") & (pass_outcome.isna())).astype(int)

    df["is_key_pass"] = (df.get("pass_shot_assist", False).fillna(False) & df["type"].eq("Pass")).astype(int)
    df["is_assist"]   = (df.get("pass_goal_assist", False).fillna(False) & df["type"].eq("Pass")).astype(int)

    dribble_outcome = df.get("dribble_outcome", pd.Series([np.nan]*len(df))).astype(str).str.lower()
    df["is_dribble_cmp"] = ((df["type"] == "Dribble") & (dribble_outcome.str.contains("complete"))).astype(int)

    duel_outcome = df.get("duel_outcome", pd.Series([np.nan]*len(df))).astype(str).str.lower()
    df["is_duel_won"] = ((df["type"] == "Duel") & (duel_outcome.str.contains("won"))).astype(int)

    inter_outcome = df.get("interception_outcome", pd.Series([np.nan]*len(df))).astype(str).str.lower()
    df["is_interception"] = ((df["type"] == "Interception") & (inter_outcome.str.contains("won|success"))).astype(int)

    block_cols = [c for c in df.columns if c.startswith("block_")]
    if block_cols:
        blocks_any = df[block_cols].fillna(False).any(axis=1)
        df["is_block"] = (blocks_any & df["type"].eq("Block")).astype(int)
    else:
        df["is_block"] = 0

    if "type" in df.columns and df["type"].eq("Clearance").any():
        df["is_clearance"] = df["type"].eq("Clearance").astype(int)
    else:
        clr_cols = [c for c in df.columns if c.startswith("clearance_")]
        if clr_cols:
            cl_any = df[clr_cols].fillna(False).any(axis=1)
            df["is_clearance"] = cl_any.astype(int)
        else:
            df["is_clearance"] = 0

    # grouped totals
    shots          = df.groupby(grp)["is_shot"].sum(min_count=1).rename("shots")
    goals          = df.groupby(grp)["is_goal"].sum(min_count=1).rename("goals")
    xg             = df.groupby(grp)["_xg"].sum(min_count=1).rename("xg")

    passes_att     = df.groupby(grp)["is_pass"].sum(min_count=1).rename("passes_att")
    passes_cmp     = df.groupby(grp)["is_pass_cmp"].sum(min_count=1).rename("passes_cmp")
    key_passes     = df.groupby(grp)["is_key_pass"].sum(min_count=1).rename("key_passes")
    assists        = df.groupby(grp)["is_assist"].sum(min_count=1).rename("assists")
    dribbles_cmp   = df.groupby(grp)["is_dribble_cmp"].sum(min_count=1).rename("dribbles_cmp")
    duels_won      = df.groupby(grp)["is_duel_won"].sum(min_count=1).rename("duels_won")
    interceptions  = df.groupby(grp)["is_interception"].sum(min_count=1).rename("interceptions")
    blocks         = df.groupby(grp)["is_block"].sum(min_count=1).rename("blocks")
    clearances     = df.groupby(grp)["is_clearance"].sum(min_count=1).rename("clearances")

    # OBV
    obv_total   = df.get("obv_total_net", pd.Series([np.nan]*len(df)))
    obv_for     = df.get("obv_for_net",   pd.Series([np.nan]*len(df)))
    obv_against = df.get("obv_against_net", pd.Series([np.nan]*len(df)))

    obv_total_g    = df.assign(_obv_total=obv_total).groupby(grp)["_obv_total"].sum(min_count=1).rename("obv_total")
    obv_for_g      = df.assign(_obv_for=obv_for).groupby(grp)["_obv_for"].sum(min_count=1).rename("obv_for")
    obv_against_g  = df.assign(_obv_against=obv_against).groupby(grp)["_obv_against"].sum(min_count=1).rename("obv_against")

    # xA (needs shot_key_pass_id, id, xg_col)
    xA = pd.Series(dtype=float, name="xA")
    if {"shot_key_pass_id","id",xg_col}.issubset(df.columns):
        shots_kp  = df.loc[df["type"].eq("Shot"), ["id","shot_key_pass_id",xg_col,"season"]].dropna(subset=["shot_key_pass_id"])
        passes_kp = df.loc[df["type"].eq("Pass"), ["id","player_id","player","team","season"]]
        xA_join   = shots_kp.merge(passes_kp, left_on=["shot_key_pass_id","season"], right_on=["id","season"], how="left")
        xA        = xA_join.groupby(["season","team","player_id","player"])[xg_col].sum(min_count=1).rename("xA")

    pieces = [shots, goals, xg, passes_att, passes_cmp, key_passes, assists,
              dribbles_cmp, duels_won, interceptions, blocks, clearances,
              obv_total_g, obv_for_g, obv_against_g]
    if not xA.empty:
        pieces.append(xA)

    agg = pd.concat(pieces, axis=1).fillna(0)

    # pass%
    agg["pass_pct"] = np.where(agg["passes_att"] > 0,
                               100.0 * agg["passes_cmp"] / agg["passes_att"],
                               np.nan)

    # nice dtypes for counts
    for c in ["shots","goals","passes_att","passes_cmp","key_passes","assists",
              "dribbles_cmp","duels_won","interceptions","blocks","clearances"]:
        if c in agg.columns:
            agg[c] = agg[c].astype("Int64")

    return agg.reset_index().sort_values(["season","team","player"]).reset_index(drop=True)

# =========================================================
# 4) Minutes & games per season-team-player; per90s
# =========================================================
def player_team_season_minutes_games(data_all: pd.DataFrame) -> pd.DataFrame:
    """
    Returns minutes and games by (season, team, player_id, player):
      minutes = sum of estimated minutes_played
      games   = #matches with minutes_played > 0
    """
    mp = estimate_minutes_by_match_player(data_all)
    grp = ["season","team","player_id","player"]

    minutes = mp.groupby(grp)["minutes_played"].sum(min_count=1).rename("minutes")
    mp["played_flag"] = (mp["minutes_played"] > 0).astype(int)
    games = mp.groupby(grp)["played_flag"].sum(min_count=1).rename("games").astype("Int64")

    out = pd.concat([minutes, games], axis=1).reset_index()
    out[["minutes"]] = out[["minutes"]].fillna(0.0)
    out[["games"]]   = out[["games"]].fillna(0)
    return out

def add_per90(agg_counts: pd.DataFrame, minutes_games: pd.DataFrame,
              per90_cols: Optional[List[str]] = None) -> pd.DataFrame:
    """
    Merge totals with minutes/games and compute per90 for selected columns.
    """
    if per90_cols is None:
        per90_cols = [
            "shots","goals","xg","passes_att","passes_cmp","key_passes","assists","xA",
            "dribbles_cmp","duels_won","interceptions","blocks","clearances",
            "obv_total","obv_for","obv_against"
        ]
    df = agg_counts.merge(minutes_games, on=["season","team","player_id","player"], how="left")
    df["minutes"] = df["minutes"].fillna(0.0)
    denom = df["minutes"].replace(0, np.nan) / 90.0
    for c in per90_cols:
        if c in df.columns:
            df[f"{c}_per90"] = df[c] / denom
    return df

# =========================================================
# 5) Full pipeline
# =========================================================
def build_player_team_season_table(data_all: pd.DataFrame) -> pd.DataFrame:
    """
    Pipeline:
      1) (Optional) add_xy_all(data_all)  # not required for aggregations
      2) Aggregate season-team-player totals
      3) Estimate minutes & games season-team-player
      4) Compute per90s
    Returns one row per (season, team, player_id, player).
    """
    agg = aggregate_player_team_season_counts(data_all)
    mg  = player_team_season_minutes_games(data_all)
    out = add_per90(agg, mg)
    # Friendly column order
    first = ["season","team","player_id","player","games","minutes"]
    others = [c for c in out.columns if c not in first]
    return out[first + others]

# =========================================================
# Example usage (uncomment and run in your notebook):
# =========================================================
# data_all_xy = add_xy_all(data_all)                    # if you want x/y/z columns
# table = build_player_team_season_table(data_all)      # main seasonal table
# display(table.head())
```

```python
from typing import List, Sequence, Optional

def _detect_per90_cols(df: pd.DataFrame) -> List[str]:
    return sorted([c for c in df.columns if c.endswith("_per90")])

def _season_sort_key(s):
    # If season is "YYYY" string or int-like; fallback lexicographic if needed
    try:
        return int(str(s).split("-")[0])  # handles "2023", "2023-24", etc.
    except:
        return s

def recency_weights_k4() -> List[float]:
    """
    Default recency weights for up to 4 seasons (oldest -> newest).
    Emphasize last two seasons heavily.
    """
    return [0.05, 0.10, 0.20, 0.65]

def _effective_weights(weights: Sequence[float], n: int) -> np.ndarray:
    """Trim/normalize weights to the last n seasons."""
    w = np.array(weights[-n:], dtype=float)
    w = w / w.sum() if w.sum() > 0 else np.ones(n) / n
    return w

def _minutes_scaler(minutes: pd.Series, floor: float = 450.0, cap: float = 2700.0, power: float = 0.5) -> np.ndarray:
    """
    Convert minutes into a stabilization factor in [~0,1.0+] (sqrt-like).
    - floor: minutes below this shrink a lot; above this less.
    - cap:   avoid exploding influence from massive minutes.
    - power: 0.5 = sqrt; 1.0 = linear; 0.0 = ignore minutes.
    Returns an array the same length as minutes.
    """
    m = minutes.fillna(0).clip(lower=0, upper=cap).astype(float).values
    # normalize to [0, 1] using floor as reference, then raise to power
    z = np.clip(m / floor, 0.0, cap / floor)  # >= 1 means "enough sample"
    return np.power(z, power)

def build_recency_weighted_per90(
    seasonal_table: pd.DataFrame,
    keys: List[str] = ["player_id", "player"],   # make it ["player_id","player","team"] if you want team-specific
    K: int = 4,
    recency_weights: Optional[Sequence[float]] = None,
    use_minutes_scaler: bool = True,
    minutes_col: str = "minutes",
) -> pd.DataFrame:
    """
    Returns one row per 'keys' with recency-weighted per90 profile using up to the last K seasons available.
    - 'seasonal_table' is the output from build_player_team_season_table(data_all)
    - 'keys' define the identity you want to aggregate on (player-level or player-team-level)
    - 'recency_weights' oldest->newest; default = [0.05, 0.15, 0.30, 0.50]
    - 'use_minutes_scaler' multiplies recency weights by a minutes-based factor to de-emphasize tiny samples
    """
    if recency_weights is None:
        recency_weights = recency_weights_k4()

    # Verify columns exist
    per90_cols = _detect_per90_cols(seasonal_table)
    if not per90_cols:
        raise ValueError("No *_per90 columns found. Make sure you ran add_per90(...) earlier.")
    required = set(keys + ["season"])
    missing = required - set(seasonal_table.columns)
    if missing:
        raise ValueError(f"seasonal_table is missing columns required for grouping: {missing}")

    if minutes_col not in seasonal_table.columns:
        raise ValueError(f"Minutes column '{minutes_col}' not found in seasonal_table.")

    # We will aggregate group-by keys
    df = seasonal_table.copy()
    # Force season ordering within each group
    df["_season_key"] = df["season"].apply(_season_sort_key)

    out_rows = []
    for gvals, gdf in df.groupby(keys, dropna=False):
        # Sort seasons ascending and keep only the last K
        gdf = gdf.sort_values("_season_key", ascending=True)
        gdf = gdf.tail(K)

        n = len(gdf)
        if n == 0:
            continue

        # Build recency weights aligned (oldest -> newest) then normalize for n
        base_w = _effective_weights(recency_weights, n)

        # Minutes scaler per row (to down-weight tiny samples)
        if use_minutes_scaler:
            msc = _minutes_scaler(gdf[minutes_col])
            w = base_w * msc
            if w.sum() > 0:
                w = w / w.sum()
            else:
                w = base_w
        else:
            w = base_w

        # Weighted average for each per90 column
        weighted = {}
        for c in per90_cols:
            vals = gdf[c].astype(float).values
            # If all NaN, result NaN
            if np.isnan(vals).all():
                weighted[c.replace("_per90", "_per90_rw")] = np.nan
            else:
                # mask NaNs to avoid contaminating weights
                mask = ~np.isnan(vals)
                if mask.sum() == 0:
                    weighted[c.replace("_per90", "_per90_rw")] = np.nan
                else:
                    ww = w[mask]
                    ww = ww / ww.sum() if ww.sum() > 0 else np.ones(mask.sum()) / mask.sum()
                    weighted[c.replace("_per90", "_per90_rw")] = np.dot(vals[mask], ww)

        # Keep helpers: seasons_used, minutes_sum, games_sum, last_season_* for reference
        seasons_used = list(gdf["season"].astype(str))
        weighted["seasons_used"] = "|".join(seasons_used)
        weighted["n_seasons"] = n
        weighted["minutes_sum"] = float(gdf[minutes_col].fillna(0).sum())
        if "games" in gdf.columns:
            weighted["games_sum"] = int(gdf["games"].fillna(0).sum())
        # last season snapshot for quick reference
        last_row = gdf.iloc[-1].to_dict()
        weighted["last_season"] = str(last_row.get("season"))
        weighted["last_minutes"] = float(last_row.get(minutes_col, np.nan))

        # Compose final row
        if not isinstance(gvals, tuple):
            gvals = (gvals,)
        row = dict(zip(keys, gvals))
        row.update(weighted)
        out_rows.append(row)

    out = pd.DataFrame(out_rows)

    # Optional: sort nicely
    sort_cols = [k for k in keys]
    if "last_season" in out.columns:
        sort_cols += ["last_season"]
    return out.sort_values(sort_cols, ascending=True).reset_index(drop=True)
```

```python
# ==============================================
# Baseline dynamic-xG framework (minimal, swappable)
# ==============================================
# Requirements: pandas, numpy, scikit-learn

from dataclasses import dataclass
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV
from sklearn.pipeline import Pipeline

# ----------------------------------------------------------
# 0) Recency-weighted per90 profiles (from your previous step)
# ----------------------------------------------------------
def _detect_per90_cols(df: pd.DataFrame) -> List[str]:
    return sorted([c for c in df.columns if c.endswith("_per90")])

def _season_sort_key(s):
    try:
        return int(str(s).split("-")[0])
    except Exception:
        return s

def recency_weights_k4() -> List[float]:
    return [0.05, 0.15, 0.30, 0.50]  # oldest -> newest

def _effective_weights(weights: Sequence[float], n: int) -> np.ndarray:
    w = np.array(weights[-n:], dtype=float)
    w = w / w.sum() if w.sum() > 0 else np.ones(n)/n
    return w

def _minutes_scaler(minutes: pd.Series, floor: float = 450.0, cap: float = 2700.0, power: float = 0.5) -> np.ndarray:
    m = minutes.fillna(0).clip(lower=0, upper=cap).astype(float).values
    z = np.clip(m / floor, 0.0, cap / floor)
    return np.power(z, power)

def build_recency_weighted_per90(
    seasonal_table: pd.DataFrame,
    keys: List[str] = ["player_id","player"],  # or ["player_id","player","team"] for team-aware
    K: int = 4,
    recency_weights: Optional[Sequence[float]] = None,
    use_minutes_scaler: bool = True,
    minutes_col: str = "minutes",
    suffix: str = "_per90_rw"
) -> pd.DataFrame:
    if recency_weights is None:
        recency_weights = recency_weights_k4()

    per90_cols = _detect_per90_cols(seasonal_table)
    if not per90_cols:
        raise ValueError("No *_per90 columns found in seasonal_table.")

    req = set(keys + ["season", minutes_col])
    miss = req - set(seasonal_table.columns)
    if miss:
        raise ValueError(f"Missing columns in seasonal_table: {miss}")

    df = seasonal_table.copy()
    df["_season_key"] = df["season"].apply(_season_sort_key)

    out_rows = []
    for gvals, gdf in df.groupby(keys, dropna=False):
        gdf = gdf.sort_values("_season_key").tail(K)
        n = len(gdf)
        if n == 0:
            continue

        base_w = _effective_weights(recency_weights, n)
        if use_minutes_scaler:
            msc = _minutes_scaler(gdf[minutes_col])
            w = base_w * msc
            w = w / w.sum() if w.sum() > 0 else base_w
        else:
            w = base_w

        row = dict(zip(keys, gvals if isinstance(gvals, tuple) else (gvals,)))
        for c in per90_cols:
            vals = gdf[c].astype(float).values
            mask = ~np.isnan(vals)
            if mask.sum() == 0:
                row[c.replace("_per90", suffix)] = np.nan
            else:
                ww = w[mask]
                ww = ww/ww.sum() if ww.sum() > 0 else np.ones(mask.sum())/mask.sum()
                row[c.replace("_per90", suffix)] = float(np.dot(vals[mask], ww))

        row["seasons_used"] = "|".join(map(str, gdf["season"].tolist()))
        row["minutes_sum"] = float(gdf[minutes_col].fillna(0).sum())
        row["games_sum"] = int(gdf["games"].fillna(0).sum()) if "games" in gdf.columns else None
        row["last_season"] = str(gdf["season"].iloc[-1])
        out_rows.append(row)

    return pd.DataFrame(out_rows)

# ----------------------------------------------------------
# 1) Simple role clustering on recency-weighted per90
# ----------------------------------------------------------
@dataclass
class RoleClustererSimple:
    n_roles: int = 14
    random_state: int = 7
    model: Optional[KMeans] = None
    scaler: Optional[StandardScaler] = None
    per90_cols_: Optional[List[str]] = None

    def fit(self, rw_profiles: pd.DataFrame) -> "RoleClustererSimple":
        per90_cols = [c for c in rw_profiles.columns if c.endswith("_per90_rw")]
        if not per90_cols:
            raise ValueError("No *_per90_rw columns found. Run build_recency_weighted_per90 first.")
        X = rw_profiles[per90_cols].fillna(0.0).values
        self.per90_cols_ = per90_cols
        self.scaler = StandardScaler()
        Xs = self.scaler.fit_transform(X)
        self.model = KMeans(n_clusters=self.n_roles, random_state=self.random_state, n_init=20)
        self.model.fit(Xs)
        return self

    def predict(self, rw_profiles: pd.DataFrame) -> pd.Series:
        if self.model is None or self.scaler is None or self.per90_cols_ is None:
            raise RuntimeError("RoleClustererSimple is not fitted.")
        X = rw_profiles[self.per90_cols_].fillna(0.0).values
        Xs = self.scaler.transform(X)
        labels = self.model.predict(Xs)
        return pd.Series(labels, index=rw_profiles.index, name="role_cluster")

    def fit_predict(self, rw_profiles: pd.DataFrame, id_cols: List[str]) -> pd.DataFrame:
        self.fit(rw_profiles)
        labels = self.predict(rw_profiles)
        out = rw_profiles[id_cols].copy()
        out["role_cluster"] = labels.values
        return out

# ----------------------------------------------------------
# 2) Build match-team targets and appearances
#    (simplest v1: "on-pitch" ‚âà appeared in the match; xG per team from shots)
# ----------------------------------------------------------
def build_match_targets_and_appearances(
    events: pd.DataFrame,
    xg_col: str = "shot_statsbomb_xg"
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Returns:
      - targets: one row per (match_id, team_id/team) with:
            xg_for, xg_against, y_deltaxg = xg_for - xg_against,
            and control 'is_home' if home/away cols are present
      - appearances: mapping (match_id, team_id) -> list of player_ids that appeared

    Notes (v1 baseline):
      * Treats any player with an event in a match as 'on' (simplest first pass).
      * xG_for: sum of xg_col for that team's shots; xG_against: opponent sum.
    """
    # --- Required columns check ---
    req_cols = {"match_id", "team_id", "team", "type", "player_id"}
    missing = req_cols - set(events.columns)
    if missing:
        raise ValueError(f"events is missing required columns: {missing}")

    ev = events.copy()

    # --- Ensure xg column exists & numeric ---
    if xg_col not in ev.columns:
        ev[xg_col] = 0.0
    ev[xg_col] = pd.to_numeric(ev[xg_col], errors="coerce").fillna(0.0)

    # --- xG For by match-team (ROW filter only) ---
    shots = ev.loc[ev["type"].eq("Shot"), ["match_id", "team_id", "team", xg_col]].copy()
    xg_for = (
        shots.groupby(["match_id", "team_id", "team"], as_index=False)[xg_col]
             .sum()
             .rename(columns={xg_col: "xg_for"})
    )

    # --- xG Against = total in match - team's own ---
    xg_match = (
        xg_for.groupby("match_id", as_index=False)["xg_for"]
              .sum()
              .rename(columns={"xg_for": "xg_total_match"})
    )
    targets = xg_for.merge(xg_match, on="match_id", how="left")
    targets["xg_against"] = targets["xg_total_match"] - targets["xg_for"]
    targets["y_deltaxg"] = targets["xg_for"] - targets["xg_against"]
    targets = targets.drop(columns=["xg_total_match"])

    # --- Home control (if available) ---
    if {"home_team", "away_team"}.issubset(ev.columns):
        home_away = ev.groupby("match_id")[["home_team", "away_team"]].first().reset_index()
        targets = targets.merge(home_away, on="match_id", how="left")
        # Prefer team names match; adapt to team_id if you prefer
        targets["is_home"] = (targets["team"] == targets["home_team"]).astype(int)
        targets = targets.drop(columns=["home_team", "away_team"])
    else:
        targets["is_home"] = 0

    # --- Appearances: players per match-team ---
    # Treat any event row from a player as an appearance in v1
    appearances = (
        ev.dropna(subset=["player_id"])
          .groupby(["match_id", "team_id"])["player_id"]
          .unique()
          .reset_index()
          .rename(columns={"player_id": "player_ids"})
    )

    return targets, appearances

# ----------------------------------------------------------
# 3) Design matrix builder (players + role-mix + controls)
# ----------------------------------------------------------
@dataclass
class DesignMatrixBuilder:
    min_app_per_player: int = 5  # drop super-rare players to keep matrix sane
    role_col: str = "role_cluster"

    def build(
        self,
        targets: pd.DataFrame,
        appearances: pd.DataFrame,
        player_roles: pd.DataFrame,
        id_cols_roles: List[str] = ["player_id"]
    ) -> Tuple[pd.DataFrame, pd.Series, Dict[str, List[int]], List[str]]:
        """
        Returns:
          X: DataFrame design matrix with columns:
               - P_* player dummies
               - R_* role-mix counts
               - controls (is_home)
          y: target vector (ŒîxG)
          mapping: {"players": [player_ids order], "roles": [role ids order]}
          colnames: list of columns in X
        """
        # Merge targets with appearances
        df = targets.merge(appearances, on=["match_id","team_id"], how="left")
        df["player_ids"] = df["player_ids"].apply(lambda x: x if isinstance(x, (list, np.ndarray)) else [])

        # Map player -> role
        pr = player_roles[id_cols_roles + [self.role_col]].drop_duplicates("player_id").set_index("player_id")
        # Filter players by min appearances
        counts = df["player_ids"].explode().value_counts()
        kept_players = counts[counts >= self.min_app_per_player].index.astype(int).tolist()

        # Enumerate kept players
        player_list = sorted(kept_players)
        player_index = {pid: i for i, pid in enumerate(player_list)}
        # Roles
        role_list = sorted(pr[self.role_col].dropna().astype(int).unique().tolist())
        role_index = {rid: i for i, rid in enumerate(role_list)}

        # Build sparse-ish matrices into dense DataFrame (small leagues should be OK with ridge)
        rows = []
        for _, r in df.iterrows():
            pids = [int(p) for p in r["player_ids"] if int(p) in player_index]
            # player dummies
            row = np.zeros(len(player_list), dtype=float)
            for pid in pids:
                row[player_index[pid]] = 1.0

            # role mix
            # derive roles of participating players
            roles_for_row = pr.loc[[pid for pid in pids if pid in pr.index], self.role_col].values if len(pids) else []
            role_counts = np.zeros(len(role_list), dtype=float)
            for rid in roles_for_row:
                if int(rid) in role_index:
                    role_counts[role_index[int(rid)]] += 1.0

            # controls
            is_home = float(r.get("is_home", 0.0))

            rows.append({
                "match_id": r["match_id"],
                "team_id": r["team_id"],
                "y_deltaxg": r["y_deltaxg"],
                "is_home": is_home,
                "players": row,
                "roles": role_counts
            })

        # Assemble X, y
        X_parts = []
        colnames = []

        # Player columns
        if player_list:
            P = np.vstack([rw["players"] for rw in rows]) if rows else np.empty((0, len(player_list)))
            P_cols = [f"P_{pid}" for pid in player_list]
            X_parts.append(pd.DataFrame(P, columns=P_cols))
            colnames.extend(P_cols)

        # Role-mix columns
        if role_list:
            R = np.vstack([rw["roles"] for rw in rows]) if rows else np.empty((0, len(role_list)))
            R_cols = [f"R_{rid}" for rid in role_list]
            X_parts.append(pd.DataFrame(R, columns=R_cols))
            colnames.extend(R_cols)

        # Controls
        C = pd.DataFrame({"is_home": [rw["is_home"] for rw in rows]})
        X_parts.append(C)
        colnames.append("is_home")

        X = pd.concat(X_parts, axis=1).astype(float) if X_parts else pd.DataFrame()
        y = pd.Series([rw["y_deltaxg"] for rw in rows], name="y_deltaxg").astype(float)

        mapping = {"players": player_list, "roles": role_list}
        return X, y, mapping, colnames

# ----------------------------------------------------------
# 4) RAPM-like ridge model
# ----------------------------------------------------------
@dataclass
class RAPMSimple:
    alphas: Tuple[float, ...] = (0.1, 0.3, 1.0, 3.0, 10.0)
    model: Optional[Pipeline] = None
    coef_: Optional[pd.Series] = None
    features_: Optional[List[str]] = None

    def fit(self, X: pd.DataFrame, y: pd.Series) -> "RAPMSimple":
        if X is None or len(X) == 0:
            raise ValueError("Design matrix X is empty.")
        if y is None or len(y) == 0:
            raise ValueError("Target y is empty.")
        if len(X) != len(y):
            raise ValueError(f"X and y length mismatch: {len(X)} vs {len(y)}")

        self.features_ = list(X.columns)

        # No store_cv_values (incompatible in recent sklearn). Use cv=5.
        self.model = Pipeline([
            ("scaler", StandardScaler(with_mean=True, with_std=True)),
            ("ridge", RidgeCV(alphas=self.alphas, fit_intercept=True, cv=5))
        ])
        self.model.fit(X.values, y.values)

        # Coeffs in standardized feature space (OK for scoring via same pipeline)
        coefs = self.model.named_steps["ridge"].coef_
        self.coef_ = pd.Series(coefs, index=self.features_, name="coef_stdspace")
        return self

    def predict(self, X: pd.DataFrame) -> np.ndarray:
        if self.model is None:
            raise RuntimeError("Model not fitted.")
        # Keep column order identical to training features
        X = X.reindex(columns=self.features_, fill_value=0.0)
        return self.model.predict(X.values)

    def feature_importance(self, topn: int = 30) -> pd.DataFrame:
        if self.coef_ is None:
            raise RuntimeError("Model not fitted.")
        imp = self.coef_.abs().sort_values(ascending=False).head(topn)
        return pd.DataFrame({"feature": imp.index, "abs_coef": imp.values})

# ----------------------------------------------------------
# 5) Lineup scorer (dynamic via role-mix + player terms)
# ----------------------------------------------------------
@dataclass
class LineupScorer:
    dm_builder: DesignMatrixBuilder
    rapm: RAPMSimple
    mapping: Dict[str, List[int]]
    role_lookup: pd.Series  # player_id -> role_cluster
    feature_names: List[str]

    def _vectorize_lineup(self, player_ids: List[int], is_home: int = 1) -> pd.DataFrame:
        # Players
        P_cols = [c for c in self.feature_names if c.startswith("P_")]
        player_list = self.mapping["players"]
        row_p = np.zeros(len(P_cols), dtype=float)
        pid_to_col = {int(c.split("_",1)[1]): i for i, c in enumerate(P_cols)}
        for pid in player_ids:
            if pid in pid_to_col:
                row_p[pid_to_col[pid]] = 1.0

        # Roles
        R_cols = [c for c in self.feature_names if c.startswith("R_")]
        role_list = self.mapping["roles"]
        row_r = np.zeros(len(R_cols), dtype=float)
        rid_to_col = {int(c.split("_",1)[1]): i for i, c in enumerate(R_cols)}
        for pid in player_ids:
            rid = self.role_lookup.get(pid, None)
            if rid is not None and int(rid) in rid_to_col:
                row_r[rid_to_col[int(rid)]] += 1.0

        # Controls
        row_c = np.array([float(is_home)], dtype=float)

        # Assemble
        parts = []
        if len(P_cols):
            parts.append(pd.DataFrame([row_p], columns=P_cols))
        if len(R_cols):
            parts.append(pd.DataFrame([row_r], columns=R_cols))
        parts.append(pd.DataFrame([row_c], columns=["is_home"]))

        x = pd.concat(parts, axis=1)
        # Ensure column order = model feature order
        x = x.reindex(columns=self.feature_names, fill_value=0.0)
        return x

    def score_lineup(self, player_ids: List[int], is_home: int = 1) -> float:
        X = self._vectorize_lineup(player_ids, is_home=is_home)
        yhat = self.rapm.predict(X)[0]
        return float(yhat)

    def delta_swap(self, current: List[int], out_pid: int, in_pid: int, is_home: int = 1) -> float:
        """Return predicted Œî(ŒîxG) for swapping out_pid -> in_pid."""
        new_lineup = [pid for pid in current if pid != out_pid] + [in_pid]
        before = self.score_lineup(current, is_home=is_home)
        after  = self.score_lineup(new_lineup, is_home=is_home)
        return after - before

# ----------------------------------------------------------
# 6) End-to-end training helper
# ----------------------------------------------------------
def train_baseline_dynamic_xg(
    seasonal_table: pd.DataFrame,
    events: pd.DataFrame,
    team_aware: bool = False,
    n_roles: int = 14,
    min_app_per_player: int = 5
):
    """
    Trains the baseline:
      1) recency-weighted per90 (team-aware optional)
      2) K-Means roles
      3) match targets (ŒîxG) + appearances
      4) design matrix (players + role-mix + home)
      5) ridge model
    Returns:
      role_df: DataFrame [player_id,(player,team),role_cluster]
      rapm: fitted RAPMSimple
      scorer: LineupScorer instance ready to score arbitrary lineups
      matrix_info: dict with X, y, mapping, feature_names, targets used
    """
    # 1) recency-weighted profiles
    keys = ["player_id","player","team"] if team_aware else ["player_id","player"]
    rw = build_recency_weighted_per90(seasonal_table, keys=keys, K=4)

    # 2) roles
    rc = RoleClustererSimple(n_roles=n_roles).fit(rw)
    roles = rc.predict(rw)
    role_df = rw[keys].copy()
    role_df["role_cluster"] = roles.values

    # 3) match targets + appearances
    targets, appearances = build_match_targets_and_appearances(events)

    # 4) design matrix
    dm = DesignMatrixBuilder(min_app_per_player=min_app_per_player)
    X, y, mapping, cols = dm.build(targets, appearances, role_df, id_cols_roles=["player_id"])

    # 5) ridge model
    rapm = RAPMSimple().fit(X, y)

    # Lineup scorer
    role_lookup = role_df.drop_duplicates("player_id").set_index("player_id")["role_cluster"]
    scorer = LineupScorer(dm, rapm, mapping, role_lookup, cols)

    matrix_info = {"X": X, "y": y, "mapping": mapping, "feature_names": cols, "targets": targets}
    return role_df, rapm, scorer, matrix_info
```

```python
from typing import List, Tuple, Optional, Dict, Any

def _parse_datetime_safe(s: pd.Series) -> pd.Series:
    return pd.to_datetime(s, errors="coerce")

def _match_selector_for_team(events: pd.DataFrame, team_name: str) -> pd.DataFrame:
    """Return a small table with the team‚Äôs matches and dates to pick the latest."""
    cols = ["match_id", "team", "match_date"]
    df = events.loc[events["team"] == team_name, cols].drop_duplicates(subset=["match_id"])
    if "match_date" in df.columns:
        df["match_date_parsed"] = _parse_datetime_safe(df["match_date"])
    else:
        df["match_date_parsed"] = pd.NaT
    return df

def _extract_starting_xi_from_tactics_row(tactics_val: Any) -> List[Dict[str, Any]]:
    """
    StatsBomb 'Starting XI' event usually stores a 'tactics' dict with key 'lineup',
    which is a list of dicts containing 'player' and 'player_id' (and 'position', 'jersey_number').
    This function extracts that list (or returns []) if not available.
    """
    if isinstance(tactics_val, dict):
        # Typical shape: {'formation': '4-2-3-1', 'lineup': [ {...}, {...}, ... ]}
        if "lineup" in tactics_val and isinstance(tactics_val["lineup"], list):
            return tactics_val["lineup"]
        # Some exports might nest differently; add guards here if needed.
    return []

def _get_starting_xi_from_events(events_match_team: pd.DataFrame) -> Optional[pd.DataFrame]:
    """
    Try to read the 'Starting XI' row and unpack its 'tactics.lineup'.
    Returns a DataFrame with columns: player_id, player, position (if present), jersey_number (if present).
    If not found, returns None.
    """
    xi_rows = events_match_team.loc[events_match_team["type"] == "Starting XI"]
    if xi_rows.empty:
        return None

    # If multiple (rare), take the first
    row = xi_rows.iloc[0]
    lineup = _extract_starting_xi_from_tactics_row(row.get("tactics", {}))
    if not lineup:
        return None

    recs = []
    for p in lineup:
        recs.append({
            "player_id": p.get("player_id"),
            "player": p.get("player"),
            "position": (p.get("position") or {}).get("name") if isinstance(p.get("position"), dict) else p.get("position"),
            "jersey_number": p.get("jersey_number"),
        })
    xi_df = pd.DataFrame(recs).dropna(subset=["player_id"]).drop_duplicates("player_id")
    # Keep only 11 if somehow more were listed
    return xi_df.head(11)

def _infer_first11_by_appearance(events_match_team: pd.DataFrame) -> pd.DataFrame:
    """
    Fallback: infer the XI as the 11 players with the earliest event times in the match for this team.
    More robust if you add proper on/off parsing; for a quick test, this works.
    """
    minute = pd.to_numeric(events_match_team.get("minute", np.nan), errors="coerce").fillna(0)
    second = pd.to_numeric(events_match_team.get("second", np.nan), errors="coerce").fillna(0)
    tmin = minute + second/60.0

    df = events_match_team.assign(_tmin=tmin)
    first_seen = (df.dropna(subset=["player_id"])
                    .groupby(["player_id","player"], as_index=False)["_tmin"]
                    .min()
                    .sort_values("_tmin", ascending=True)
                    .head(11))
    first_seen = first_seen.rename(columns={"_tmin": "first_minute"})
    return first_seen[["player_id","player"]]

def get_last_starting_xi_ids(
    events: pd.DataFrame,
    team_name: str = "Am√©rica"
) -> Tuple[List[int], pd.DataFrame, Dict[str, Any]]:
    """
    Returns:
      - xi_ids: list[int] of the latest Starting XI player_ids for the team
      - xi_df:  DataFrame with player_id, player, and (if available) position, jersey_number
      - meta:   dict with 'match_id' and 'match_date' used
    Strategy:
      1) Find the latest match for team_name by match_date (fallback: max match_id).
      2) Try to parse 'Starting XI' event's 'tactics.lineup'.
      3) Fallback to first 11 by earliest event appearance.
    """
    # 1) select latest match
    mt = _match_selector_for_team(events, team_name)
    if mt.empty:
        raise ValueError(f"No matches found for team='{team_name}'. Check 'team' values in your DF.")

    if mt["match_date_parsed"].notna().any():
        latest_row = mt.sort_values("match_date_parsed").iloc[-1]
    else:
        # Fallback if dates missing: pick the numerically largest match_id
        latest_row = mt.sort_values("match_id").iloc[-1]

    match_id = latest_row["match_id"]
    match_date = latest_row.get("match_date")
    # 2) scope events to this match & team
    emt = events[(events["match_id"] == match_id) & (events["team"] == team_name)].copy()

    # Try Starting XI
    xi_df = _get_starting_xi_from_events(emt)
    if xi_df is None or xi_df.empty:
        xi_df = _infer_first11_by_appearance(emt)
        # bring some columns for display if available
        xi_df["position"] = np.nan
        xi_df["jersey_number"] = np.nan

    xi_df = (xi_df
             .dropna(subset=["player_id"])
             .drop_duplicates("player_id")
             .head(11)
             .reset_index(drop=True))

    xi_ids = xi_df["player_id"].astype(int).tolist()
    meta = {"match_id": match_id, "match_date": match_date}
    return xi_ids, xi_df, meta
```

```python
import pandas as pd
import numpy as np
from typing import Dict, List, Optional
from collections import Counter
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# =========================================================
# 0) Position mapping (granular -> grouped)
# =========================================================
POSITION_GROUP_MAP = {
    # Fullbacks & Wingbacks
    "Right Back": "Right Back",
    "Right Wing Back": "Right Back",
    "Left Back": "Left Back",
    "Left Wing Back": "Left Back",

    # Center Backs (incl. sided CBs)
    "Center Back": "Center Back",
    "Right Center Back": "Center Back",
    "Left Center Back": "Center Back",

    # Defensive Mid (incl. sided DMs)
    "Center Defensive Midfield": "Defensive Midfield",
    "Right Defensive Midfield": "Defensive Midfield",
    "Left Defensive Midfield": "Defensive Midfield",

    # Central Mid (incl. sided CMs)
    "Center Midfield": "Center Midfield",
    "Right Center Midfield": "Center Midfield",
    "Left Center Midfield": "Center Midfield",

    # Attacking Mid (incl. sided AMs)
    "Center Attacking Midfield": "Attacking Midfield",
    "Right Attacking Midfield": "Attacking Midfield",
    "Left Attacking Midfield": "Attacking Midfield",

    # Wingers & wide mids
    "Right Wing": "Right Wing",
    "Right Midfield": "Right Wing",
    "Left Wing": "Left Wing",
    "Left Midfield": "Left Wing",

    # Forwards
    "Center Forward": "Center Forward",
    "Right Center Forward": "Center Forward",
    "Left Center Forward": "Center Forward",

    # Others
    "Goalkeeper": "Goalkeeper",
    "Substitute": "Substitute",
}

ALLOWED_ORIGINAL = list(POSITION_GROUP_MAP.keys())
ALLOWED_GROUPED  = sorted(set(POSITION_GROUP_MAP.values()))

# =========================================================
# 1) Primary grouped position with ‚â•10 distinct matches
# =========================================================
def infer_primary_grouped_position(
    events: pd.DataFrame,
    keys: List[str] = ["player_id","player","team"],  # or ["player_id","player"]
    pos_col: str = "position",
    match_col: str = "match_id",
    min_games: int = 10,
    exclude_substitute: bool = True,
) -> pd.DataFrame:
    """
    Assigns a primary GROUPED position only if a player has >= min_games matches
    with that grouped position in the events. Counts DISTINCT matches per grouped position.
    Returns: DataFrame[keys + ["primary_grouped_pos","games_in_pos"]]
    """
    req = set(keys + [pos_col, match_col])
    missing = req - set(events.columns)
    if missing:
        raise ValueError(f"events missing required columns: {missing}")

    df = events[keys + [pos_col, match_col]].dropna(subset=[pos_col, match_col]).copy()
    # keep only known granular positions
    df = df[df[pos_col].isin(ALLOWED_ORIGINAL)]
    if df.empty:
        raise ValueError("No rows with allowed positions found in events.")

    # map to grouped
    df["grouped_pos"] = df[pos_col].map(POSITION_GROUP_MAP)

    # optionally drop 'Substitute'
    if exclude_substitute:
        df = df[df["grouped_pos"] != "Substitute"]

    # count DISTINCT matches per (keys, grouped_pos)
    # if a player had multiple rows in a match at that position, count once
    gp = (df.drop_duplicates(subset=keys + ["grouped_pos", match_col])
            .groupby(keys + ["grouped_pos"], as_index=False)[match_col]
            .nunique()
            .rename(columns={match_col: "games_in_pos"}))

    # filter by threshold
    gp = gp[gp["games_in_pos"] >= min_games]
    if gp.empty:
        raise ValueError(f"No player-position combos reached >= {min_games} games.")

    # choose the grouped_pos with max games for each keys (tie-breaker = first)
    gp = gp.sort_values(["games_in_pos"], ascending=False)
    prim = gp.loc[gp.groupby(keys)["games_in_pos"].idxmax()].reset_index(drop=True)
    prim = prim.rename(columns={"grouped_pos": "primary_grouped_pos"})
    return prim  # keys + ["primary_grouped_pos","games_in_pos"]

# =========================================================
# 2) Per-position clustering (within grouped positions)
# =========================================================
def _detect_per90_rw_cols(df: pd.DataFrame) -> List[str]:
    return sorted([c for c in df.columns if c.endswith("_per90_rw")])

def _top_feature_label(centroid: np.ndarray, feature_names: List[str], k_top: int = 3) -> str:
    idx = np.argsort(np.abs(centroid))[::-1][:k_top]
    parts = []
    for j in idx:
        sign = "+" if centroid[j] >= 0 else "-"
        parts.append(f"{sign}{feature_names[j]}")
    return ", ".join(parts)

def cluster_players_per_grouped_position(
    rw_profiles: pd.DataFrame,
    prim_pos: pd.DataFrame,
    keys: List[str] = ["player_id","player","team"],  # or ["player_id","player"]
    n_clusters_by_pos: Optional[Dict[str, int]] = None,
    random_state: int = 7,
    min_group_size_for_kmeans: int = 6,
) -> pd.DataFrame:
    """
    K-Means per GROUPED position using *_per90_rw features.
    Returns: DataFrame[keys + ["primary_grouped_pos","pos_cluster_id","pos_cluster_label"]]
    """
    per90_cols = _detect_per90_rw_cols(rw_profiles)
    if not per90_cols:
        raise ValueError("No *_per90_rw columns in rw_profiles.")

    # default number of clusters per grouped position
    default_k = {
        "Goalkeeper": 3,
        "Right Back": 4, "Left Back": 4,
        "Center Back": 4,
        "Defensive Midfield": 4,
        "Center Midfield": 5,
        "Attacking Midfield": 4,
        "Right Wing": 4, "Left Wing": 4,
        "Center Forward": 4,
    }
    if n_clusters_by_pos is None:
        n_clusters_by_pos = default_k

    # join primary grouped position onto rw_profiles (only those who passed ‚â•min_games)
    df = rw_profiles.merge(prim_pos, on=keys, how="inner")  # inner: keep only qualified players
    if df.empty:
        raise ValueError("No overlap between RW profiles and primary grouped positions (‚â•min_games).")

    results = []
    for pos, g in df.groupby("primary_grouped_pos", dropna=False):
        g = g.copy()
        X = g[per90_cols].astype(float).fillna(0.0).values

        if len(g) < min_group_size_for_kmeans:
            labels = np.zeros(len(g), dtype=int)
            label_texts = ["small-group"] * len(g)
        else:
            k = int(n_clusters_by_pos.get(pos, 3))
            k = max(1, min(k, len(g)))
            scaler = StandardScaler()
            Xs = scaler.fit_transform(X)
            km = KMeans(n_clusters=k, random_state=random_state, n_init=20)
            labels = km.fit_predict(Xs)
            cents = km.cluster_centers_
            label_texts = [_top_feature_label(cents[j], per90_cols, k_top=3) for j in labels]

        sub = g[keys + ["primary_grouped_pos","games_in_pos"]].copy()
        sub["pos_cluster_id"] = labels
        sub["pos_cluster_label"] = [f"{pos} | {t}" for t in label_texts]
        results.append(sub)

    clusters = pd.concat(results, axis=0).reset_index(drop=True)
    return clusters  # keys + ["primary_grouped_pos","games_in_pos","pos_cluster_id","pos_cluster_label"]

# =========================================================
# 3) Convenience: Center Forward table (candidates for swaps)
# =========================================================
def grouped_position_table(
    clusters_df: pd.DataFrame,
    grouped_position: str = "Center Forward",
) -> pd.DataFrame:
    cols = [c for c in ["player_id","player","team","primary_grouped_pos","games_in_pos",
                        "pos_cluster_id","pos_cluster_label"]
            if c in clusters_df.columns]
    out = (clusters_df.loc[clusters_df["primary_grouped_pos"] == grouped_position, cols]
           .sort_values(["pos_cluster_id","player"], na_position="last")
           .reset_index(drop=True))
    return out
```

```python

# =========================================================
# 4) Swap simulation over CF candidates
# =========================================================
def simulate_cf_swaps(
    clusters_df: pd.DataFrame,
    scorer,                 # LineupScorer (from the model you trained)
    xi: List[int],          # current XI player_ids
    out_pid: int,           # current CF to swap out
    mapping_filter: bool = True  # keep only players known by the RAPM mapping
) -> pd.DataFrame:
    # CF candidates
    st_table = grouped_position_table(clusters_df, grouped_position="Center Midfield")
    pids = st_table["player_id"].astype(int).tolist()

    # Optional: only players the model knows (have P_* columns)
    if mapping_filter and hasattr(scorer, "mapping") and "players" in scorer.mapping:
        known = set(map(int, scorer.mapping["players"]))
        pids = [p for p in pids if p in known]

    swap_results = []
    for p in pids:
        try:
            delta = scorer.delta_swap(current=xi, out_pid=out_pid, in_pid=p, is_home=1)
        except Exception:
            delta = np.nan
        swap_results.append({"player_id": p, "delta_xg": delta})

    swap_df = pd.DataFrame(swap_results)
    result = st_table.merge(swap_df, on="player_id", how="left").sort_values(
        "delta_xg", ascending=False
    ).reset_index(drop=True)
    return result
```

```python

from collections import defaultdict

# --- same grouped mapping as before ---
POSITION_GROUP_MAP = {
    "Right Back": "Right Back", "Right Wing Back": "Right Back",
    "Left Back": "Left Back",   "Left Wing Back": "Left Back",
    "Center Back": "Center Back", "Right Center Back": "Center Back", "Left Center Back": "Center Back",
    "Center Defensive Midfield": "Defensive Midfield", "Right Defensive Midfield": "Defensive Midfield", "Left Defensive Midfield": "Defensive Midfield",
    "Center Midfield": "Center Midfield", "Right Center Midfield": "Center Midfield", "Left Center Midfield": "Center Midfield",
    "Center Attacking Midfield": "Attacking Midfield", "Right Attacking Midfield": "Attacking Midfield", "Left Attacking Midfield": "Attacking Midfield",
    "Right Wing": "Right Wing", "Right Midfield": "Right Wing",
    "Left Wing": "Left Wing",   "Left Midfield": "Left Wing",
    "Center Forward": "Center Forward", "Right Center Forward": "Center Forward", "Left Center Forward": "Center Forward",
    "Goalkeeper": "Goalkeeper",
    "Substitute": "Substitute",
}

ALLOWED_ORIGINAL = set(POSITION_GROUP_MAP.keys())

def top_positions_grouped(
    events: pd.DataFrame,
    keys: List[str] = ["player_id","player","team"],   # or ["player_id","player"]
    pos_col: str = "position",
    match_col: str = "match_id",
    date_col: str = "match_date",                      # used for tie-break if present
    min_games: int = 10,
    top_k: int = 3,
    exclude_substitute: bool = True,
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Returns:
      wide_df: one row per player (keys) with columns:
               primary_pos, primary_games, secondary_pos, secondary_games, third_pos, third_games
      long_df: per player-position summary with columns:
               keys + [grouped_pos, games, last_date]

    Logic:
      - Map granular positions -> grouped positions (POSITION_GROUP_MAP).
      - Count DISTINCT matches per (player, grouped_pos).
      - Keep only grouped_pos with games >= min_games.
      - Rank positions by games desc, then by most recent date (if available).
      - Take top_k (default 3) to fill primary/secondary/third.
    """
    req = set(keys + [pos_col, match_col])
    missing = req - set(events.columns)
    if missing:
        raise ValueError(f"events missing required columns: {missing}")

    df = events[keys + [pos_col, match_col] + ([date_col] if date_col in events.columns else [])].copy()
    df = df.dropna(subset=[pos_col, match_col])
    # keep only known granular positions
    df = df[df[pos_col].isin(ALLOWED_ORIGINAL)]
    if df.empty:
        raise ValueError("No rows with allowed original positions found in events.")

    # map to grouped
    df["grouped_pos"] = df[pos_col].map(POSITION_GROUP_MAP)
    # optionally drop Substitute
    if exclude_substitute:
        df = df[df["grouped_pos"] != "Substitute"]

    # unique match appearances per player & grouped position
    dedup_cols = keys + ["grouped_pos", match_col]
    df_dedup = df.drop_duplicates(subset=dedup_cols)

    # count games
    games = (df_dedup
             .groupby(keys + ["grouped_pos"], as_index=False)[match_col]
             .nunique()
             .rename(columns={match_col: "games"}))

    # last date for tie-break (if available)
    if date_col in df.columns:
        # get last date per (player, grouped_pos)
        tmp = df.copy()
        tmp[date_col] = pd.to_datetime(tmp[date_col], errors="coerce")
        last_date = (tmp.sort_values(date_col)
                        .dropna(subset=[date_col])
                        .groupby(keys + ["grouped_pos"], as_index=False)[date_col]
                        .max()
                    )
        summary = games.merge(last_date, on=keys + ["grouped_pos"], how="left")
    else:
        summary = games.copy()
        summary[date_col] = pd.NaT

    # filter by threshold
    summary = summary[summary["games"] >= min_games]
    if summary.empty:
        # Return empty results but with expected columns
        wide_cols = keys + [
            "primary_pos","primary_games","secondary_pos","secondary_games","third_pos","third_games"
        ]
        return pd.DataFrame(columns=wide_cols), summary

    # sort within player by games desc, then last_date desc (if exists)
    summary = summary.sort_values(keys + ["games", date_col], ascending=[True]*len(keys) + [False, False])

    # take top_k per player
    def topk_rows(g: pd.DataFrame) -> pd.Series:
        g_sorted = g.sort_values(["games", date_col], ascending=[False, False])
        # pad with Nones if fewer than top_k
        pos_list = g_sorted["grouped_pos"].tolist()[:top_k]
        game_list = g_sorted["games"].tolist()[:top_k]
        pos_list += [None] * (top_k - len(pos_list))
        game_list += [np.nan] * (top_k - len(game_list))
        out = {}
        names = ["primary","secondary","third"][:top_k]
        for i, nm in enumerate(names):
            out[f"{nm}_pos"] = pos_list[i]
            out[f"{nm}_games"] = game_list[i]
        return pd.Series(out)

    wide = (summary.groupby(keys, as_index=False)
                  .apply(topk_rows)
                  .reset_index()
                  .drop(columns=["level_0"], errors="ignore"))

    # Nice ordering: keys then primary/secondary/third
    ordered_cols = keys + ["primary_pos","primary_games","secondary_pos","secondary_games","third_pos","third_games"]
    wide = wide.reindex(columns=ordered_cols)

    # long_df for transparency/debug
    long_df = summary[keys + ["grouped_pos","games", date_col]].rename(columns={date_col: "last_date"}).reset_index(drop=True)

    return wide, long_df
```
